```python
import os

directory = "/Users/ishananand/.cache/kagglehub/datasets/prahladmehandiratta/cervical-cancer-largest-dataset-sipakmed/versions/1/"
items = os.listdir(directory)

for item in items:
    image_dir = os.path.join(directory, item, item, "CROPPED")
    
    if os.path.isdir(image_dir):
        images = [f for f in os.listdir(image_dir) if f.lower().endswith('.bmp')]
        print(f"Found {len(images)} .bmp files in {image_dir}")
        
        # for img in images:
        #     print(img)

```

```python

dest_dir = "/Users/ishananand/Cervical-Cancer-Detection-Using-Vision-Transformer/data"  # <-- change this to your desired location

# Create the destination base directory if it doesn't exist
os.makedirs(dest_dir, exist_ok=True)

for item in os.listdir(directory):
    cropped_path = os.path.join(directory, item, item, "CROPPED")
    
    if os.path.isdir(cropped_path):
        bmp_files = [f for f in os.listdir(cropped_path) if f.lower().endswith('.bmp')]
        
        # Create class folder in destination
        class_dest_dir = os.path.join(dest_dir, item)
        os.makedirs(class_dest_dir, exist_ok=True)
        
        # Copy each .bmp file
        for bmp_file in bmp_files:
            src_file = os.path.join(cropped_path, bmp_file)
            dst_file = os.path.join(class_dest_dir, bmp_file)
            shutil.copy2(src_file, dst_file)  # copy2 preserves metadata
            print(f"Copied {bmp_file} to {class_dest_dir}")
```
